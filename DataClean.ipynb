```python
import os
import time
from PIL import Image
import psutil
```

```python
Datasets="Datasets"
files=os.listdir(Datasets)
```

```python
print(files)
for item in files:
    file=os.listdir(Datasets+"/"+item)
    for image in file:
        print(Datasets+"/"+item+"/"+image)
        try:
            img = Image.open(Datasets+"/"+item+"/"+image)
            img.show()
#             time.sleep(0.2)
            for proc in psutil.process_iter():
                if proc.name() == "display":
                    proc.kill()
        except Exception:
            pass
```
